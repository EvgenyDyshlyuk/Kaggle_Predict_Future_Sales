### 1.1 Load Data

#### Input Data Folder:
The input data is in the 'input' folder of this directory

#### Data Description:  
You are provided with daily historical sales data. The task is to forecast the total amount of products sold in every shop for the test set. Note that the list of shops and products slightly changes every month. Creating a robust model that can handle such situations is part of the challenge.

#### Files description:  
- sales_train.csv - the training set. Daily historical data from January 2013 to October 2015.  
- test.csv - the test set. You need to forecast the sales for these shops and products for November 2015.  
- sample_submission.csv - a sample submission file in the correct format. 
- items.csv - supplemental information about the items/products.  
- item_categories.csv  - supplemental information about the items categories.  
- shops.csv- supplemental information about the shops.  

#### Data fields:
- ID - an Id that represents a (Shop, Item) tuple within the test set
- shop_id - unique identifier of a shop
- item_id - unique identifier of a product
- item_category_id - unique identifier of item category
- item_cnt_day - number of products sold. You are predicting a monthly amount of this measure
- item_price - current price of an item
- date - date in format dd/mm/yyyy
- date_block_num - a consecutive month number, used for convenience. January 2013 is 0, February 2013 is 1,..., October 2015 is 33
- item_name - name of item
- shop_name - name of shop
- item_category_name - name of item category

```python
#Load data from 'input' folder in the current directory
train   = pd.read_csv('input/sales_train.csv')
items   = pd.read_csv('input/items.csv')
cats    = pd.read_csv('input/item_categories.csv')
shops   = pd.read_csv('input/shops.csv')
test    = pd.read_csv('input/test.csv')
sample  = pd.read_csv('input/sample_submission.csv')

# Show the Loaded Data. # display() allows to output multiple dataframes in one cell
display('train',   train.shape,  train.head(),
        'items',   items.shape,  items.head(),
        'cats',    cats.shape,   cats.head(),
        'shops',   shops.shape,  shops.head(),
        'test',    test.shape,   test.head(),
        'sample',  sample.shape, sample.head()) 
```

```python
test = test.set_index('ID') #Set index to ID. This way we do not need to drop ID column every time in future calculations
```

## Train vs Test

```python
# Let's see the differences between train and test sets
display(train.head(3), train.tail(3),train.nunique(),
        test.head(3), test.tail(3),test.nunique())

print('Number of item_id in train but not in test:', len(set(train.item_id)-set(test.item_id)))
print('Number of item_id in test but not in train:', len(set(test.item_id)-set(train.item_id)))
```

#### We observe differences between train and test:
- 60 shops in train, while only 42 in test (other shops were closed)
- 21807 items_id in train, only 5100 items are in test
- 5100 items * 42 shops = 214200 pairs == len test set. So the test set is predictions of sales for 5100 items in 42 shops.

- 17070 items are present in train but not in test
- 4737 items present in both train and test
- 363 items in test are new compared to the train

- test set does not contain 'item_price column'

```python
# How many unique items is sold (at least one sale) every month?
for i in range(34):
    print('month ##:',i, '   unique items sold:', train[train.date_block_num == i].nunique().item_id)
```

We can see that in the last month of train - 5413 items were sold, which is very close to test set number of items. Does it mean that the test set contains only items that were sold at least once during this test month? Possibly yes: 5100 items were sold during test month and we are to predict sales for those items only.

```python
# How many absolutely new unique items is sold (at least one sale) every month?
for i in range(34):
    print('month ##:',i, 'New items:', len(set(train[train.date_block_num == i].item_id)-set(train[train.date_block_num < i].item_id)))
```

Number of new items introduced and sold in the previous months is close to number of new items introduced (and sold?) in the test month: 363

```python
# Lets see sales for one month - we see 7 days weekly variation - probably adding weekdays as a feature is important for training
groupby_and_plot(train[train.date_block_num == 33], ['date'], 'item_cnt_day')
```

There is daily trend in sales - low sales on Monday, highest sales on Weekends. Probably adding number of Mondays, Tuesdays, etc. as features in the particular month would help. Holidays also show higher sales - better take this into account.

```python
# Let us see sales distribution per months
groupby_and_plot(train, ['date_block_num'], 'item_cnt_day')
```

- We clearly see a pattern here - overall negative trend in cumulative sales, with a 12 month period sinusoidal - year cycle. Peak sales - December, low sales on summer months. 
- We see standard deviation in sales increase significantly for months 32,33 (why is this?). Predictions using data from those two months can be more difficult to make.

```python
# Let's see the sales per shop
groupby_and_plot(train, ['shop_id'], 'item_cnt_day')
```

- 4 shops: 9, 12, 20 and 55 demonstrate unusual behaviour, let us see why?

```python
# Let us plot sales per shop over time. We will use red color title for those shops, that are not present in test set.
fig = plt.figure(figsize=(30,36))
for i in set(train.shop_id):
    ts=train[train.shop_id == i].groupby(['date_block_num'])['item_cnt_day'].sum()
    plt.subplot(10, 6, i+1)
    plt.bar(ts.index, ts.values)
    plt.xlim((0, 33))
    plt.ylim(0, 12000)
    if i in set(test.shop_id):
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'k')
    else: 
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'r')
plt.show()
```

- We need to predict sales only for the shops which were not closed before :)

```python
# We see that data for some shops was split to two parts and shop names were changed (intentionally I guess), let's fix it
# Якутск Орджоникидзе, 56
train.loc[train.shop_id == 0, 'shop_id'] = 57
test.loc[test.shop_id == 0, 'shop_id'] = 57
# Якутск ТЦ "Центральный"
train.loc[train.shop_id == 1, 'shop_id'] = 58
test.loc[test.shop_id == 1, 'shop_id'] = 58
# Жуковский ул. Чкалова 39м²
train.loc[train.shop_id == 10, 'shop_id'] = 11
test.loc[test.shop_id == 10, 'shop_id'] = 11
# Now delete those shops from the shops dataframe:
shops.drop([0, 1, 10], inplace = True)

# I think it is also better to remove any data for outbound trade, 
# which is very unusual and possibly misleading (we are not going to predict the outbound trade anyway in the test)
train = train[train.shop_id != 9]
train = train[train.shop_id != 20]
# Now delete those shops from the shops dataframe:
shops.drop([9, 20], inplace = True)

# 12 and 55 are online stores - we cannot remove them because these shops are in test. These 2 shops are IT sales shops and very special compared to others.
```

```python
# plot again
fig = plt.figure(figsize=(30,36))
for i in set(train.shop_id):
    ts=train[train.shop_id == i].groupby(['date_block_num'])['item_cnt_day'].sum()
    plt.subplot(10, 6, i+1)
    plt.bar(ts.index, ts.values)
    plt.xlim((0, 33))
    plt.ylim(0, 12000)
    if i in set(test.shop_id):
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'k')
    else: 
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'r')
plt.show()
```

```python
# We see that now we have only 2 very unusual shops: ##12 and 55
groupby_and_plot(train, ['shop_id'], 'item_cnt_day')
```

```python
# Now let's plot mean values
fig = plt.figure(figsize=(30,36))
for i in set(train.shop_id):
    ts=train[train.shop_id == i].groupby(['date_block_num'])['item_cnt_day'].mean()
    plt.subplot(10, 6, i+1)
    plt.bar(ts.index, ts.values)
    plt.xlim((0, 33))
    #plt.ylim(1, 1.5)
    if i in set(test.shop_id):
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'k')
    else: 
        plt.title(str(i) +' '+ shops.shop_name[i], color = 'r')
plt.show()
```

We see that shops ## 12 and 55 demonstrate the unusual behavior especially over the last two months. Overall the mentioned unusual high STD for all shops in the last two months is due to these two shops. This is due to some big numbers of sales in the two months - these are outliers. If we clip the sales data to (0,20) range the data for those two shops looks much more normal.
